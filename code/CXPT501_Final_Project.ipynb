###**Data Cleaning & Manipulation**

Import the original Framingham Heart Study database is imported and then perform data cleaning of NaNs and outliers.

```python
#Import pandas, scipy.stats, numpy, seaborn and matplotlib.pyplot
import pandas as pd
import scipy.stats as stats
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Import original database
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1zFxWLexV3XWVf6ix7CHUrUIwvAcG0xX2')
df
```

```python
#Save original dataset to startingData directory
df.to_csv('/content/startingData/Framingham Heart Study.csv')
```

```python
#Extract data of the third exam
df_third=df.loc[df['PERIOD'] == 3]
df_third
```

```python
#Check null
df_third.isnull().sum()
```

```python
#Drop null from all column except HDLC and LDLC
df2_third=df_third.dropna(subset=['TOTCHOL', 'CIGPDAY', 'BMI', 'BPMEDS', 'HEARTRTE', 'GLUCOSE', 'educ', 'HDLC', 'LDLC'])
df2_third
```

```python
#Check null from df2
df2_third.isnull().sum()
```

```python
#Calculate the 25th percentile of BMI
qOne = df2_third['BMI'].quantile(0.25)
qOne
```

```python
#Calculate the 75th percentile of BMI
qThree = df2_third['BMI'].quantile(0.75)
qThree
```

```python
#Calculate the iqr of BMI
iqr = stats.iqr(df2_third['BMI'])
iqr
```

```python
#Calculate the maximum of BMI
maximum = qThree + (1.5 * iqr)
maximum
```

```python
#Calculate the minimum of BMI
minimum = qOne - (1.5 * iqr)
minimum
```

```python
#Calculate the outliers of BMI
iqrOutliers = df2_third['BMI'].loc[(df2_third['BMI'] < minimum) | (df2_third['BMI'] > maximum)]
iqrOutliers
```

```python
#Drop BMI outliers from df2
df3=df2_third.drop(iqrOutliers.index)
```

```python
#Check how many patients left after data cleaning
df3
```

###**Hypothesis Tests**

**Kolmogorov–Smirnov test**

To test the difference of mean BMI between patients who have cvd and patinets who don't have cvd. BMI data is continuous dependet variable, Kolmogorov–Smirnov test should be performed first to test if the sample data is normally distruibuted. 

```python
#Create subgroups for patients who have cvd and patients who don't have cvd
cvd = df3.loc[df['CVD'] == 1]
nocvd = df3.loc[df['CVD'] == 0]
```



*   H0: The data follows the chosen distribution (normal) 
*   H1: The data follows the chosen distribution (normal) 


```python
#Test if BMI is normal distributed if patients have cvd
stats.kstest(cvd['BMI'], 'norm')
```

```python
#Test if BMI is normal distributed if patients don't have cvd
stats.kstest(nocvd['BMI'], 'norm')
```

Since p-value is lower than 0.05, we reject the null hypothesis, so the BMI data of both groups are not normally distributed.

```python
#Create histogram to show distribution shape of BMI of patients with cvd
plt.figure(figsize = (10, 6), dpi = 300)

figure2=sns.histplot(data=cvd, x='BMI')
figure2.set(ylabel='')

plt.savefig("/content/figures/Figure2_Histogram1.pdf", bbox_inches='tight')
```

```python
#Create histogram to show distribution shape of BMI of patients without cvd
plt.figure(figsize = (10, 6), dpi = 300)

figure3=sns.histplot(data=nocvd, x='BMI')
figure3.set(ylabel='')

plt.savefig("/content/figures/Figure2_Histogram2.pdf", bbox_inches='tight')
```

**Mann-Whitney U Test**

Assumptions of Mann-Whitney U Test are met:


*   Patients with cvd and without cvd are two indepented groups.
*   BMI from two groups are not normal distributed.
*   The distribution of BMI for both groups have similar shape.
*   BMI is continous data.
*   Sample size of each groups is larger than 5.


So Mann-Whitney U test is used to test the different bewteem BMI in two groups.





Set significance level α=0.05.



Hypothesis:

*   H0: BMI is lower in patients with CVD than it is in patient without CVD
*   H1: BMI is higher in patients with CVD than it is in patient without CVD

```python
stat, p_value = stats.mannwhitneyu(x=cvd['BMI'], y=nocvd['BMI'], alternative = 'greater')
print(f"Mann Whitney's test statistic: {stat}") 
print(f"P-value: {p_value}") 
```

Since p<0.05, we reject the null hypothesis, meaning that BMI is statistically higher in patients with CVD than it is in patient without CVD.

**Pearson Chi-square test**

Because both sex and cvd are categorical dependent variables, the the Pearson Chi-square test is used to test the association between sex and cvd hospitalization.

Set significance level α=0.05.


*   H0: No association between sex and cvd
*   H1: Association between sex and cvd (two-sided)




```python
#Construct a contigency table for
contigencyTable = pd.crosstab(df3['SEX'], df3['CVD']).reindex([1,0],axis="columns")
contigencyTable
```

```python
#Perform person chi-square testto test the hypothesis
chi2, p, dof, expected = stats.chi2_contingency(contigencyTable)
print(f"Chi2 statistic: {chi2:.5g}")
print(f"p-value: {p:.5g}")
print(f"Dgrees of freedom: {dof}")
print("Expected frequenciesn")
```

Since p<0.05, we reject the null hypothesis, meaning there is association between sex and CVD appearance.
